### Ejercicio 2.2 (opcional):

Realiza un gráfico que contenga dos subplots. A la izquierda deberá mostrar la imagen recien cargada y a la derecha el histograma de los datos.

Tip: Para generar el histograma, considera redimensionar el array, ya que esta función trabaja con vectores en una dimensión. Esto puedes hacerlo de dos maneras:

- Forma 1:

    ```python
    filas, cols = array.shape
    
    array = array.reshape(filas * cols)
    ```

- Forma 2:
    
